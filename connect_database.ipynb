```python
server_name = "jdbc:sqlserver://34.125.61.40"
database_name = "TestDB"
url = server_name + ";" + "databaseName=" + database_name + ";"

table_name = "Inventory"
username = "SA"
password = "Passw0rd123456" # Please specify password here
```

```python
url
```

```python
synthetic_df = spark.createDataFrame([{'name':'Aekanun','gender':'M','income':1333.4}\
                                      ,{'name':'Ornanong','gender':'F','income':3334.3}])
```

```python
synthetic_df.show()
```

```python
synthetic_df.write.mode('overwrite').format("jdbc")\
        .option("url", url) \
        .option("dbtable", table_name) \
        .option("user", username) \
        .option("password", password)\
        .option("driver", "com.microsoft.sqlserver.jdbc.SQLServerDriver").save()
```

```python
jdbcDF = spark.read.format("jdbc")\
        .option("url", url) \
        .option("dbtable", table_name) \
        .option("user", username) \
        .option("password", password)\
        .option("driver", "com.microsoft.sqlserver.jdbc.SQLServerDriver")
```

```python
raw_df = jdbcDF.load()
```
